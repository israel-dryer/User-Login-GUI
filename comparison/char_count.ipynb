# Compare character counts with PySimpleGUI and Tkinter

```python
with open('gui1.txt', 'r') as f:
    sg = f.read()

with open('gui2.txt', 'r') as f:
    tk = f.read()
```

```python
sg_chars = [char.strip() for char in sg if char.strip()]
tk_chars = [char.strip() for char in tk if char.strip()]
```

```python
print('PySimpleGUI Character Count:', len(sg_chars))
print('tkinter Character Count:', len(tk_chars))
```

```python
pct_chars = (len(sg_chars) / len(tk_chars))
```

```python
print('% of Characters reduced with PySimpleGUI code: {:.2%}'.format(1-pct_chars))
```
